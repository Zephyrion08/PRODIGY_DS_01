```python
#importing libaries
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#Reading datasets
df = pd.read_csv("world_population.csv")
```

```python
# columns name change
df.columns = [each.split()[0] if (len(each.split()) > 2) else each.replace(" ", "_") for each in df.columns]
```

```python
df['Country'] = df['Country/Territory'].apply(lambda x: x.split('/')[0].strip())
```

```python
df = df.drop('Country/Territory', axis=1)
```

```python
df.head()
```

```python
df['2022_Population'].mean()
```

```python
#barplot of continent
sns.barplot(x='Continent', y='2022_Population', data=df)
```

```python
#hsitogram of continent
sns.histplot(data = df, x = 'Continent', kde = True)
```

```python
df_sorted = df.sort_values(by='Rank', ascending=True)
```

```python
df_sorted
```

```python
#country by most population in 2022
sns.barplot(y=df_sorted['Country'][:10], x=df_sorted['2022_Population'][:10],palette='coolwarm')
plt.xlabel('population')
plt.ylabel('country')
plt.show()
```

```python
#country by least population in 2022
sns.barplot(y=df_sorted['Country'][-10:], x=df_sorted['2022_Population'][-10:],palette='coolwarm')
plt.xlabel('population')
plt.ylabel('country')
plt.show()
```

```python
#counting the total no of country in each continent

sns.countplot(x='Continent', data=df)

plt.show()

```
